```python
from nltk.tokenize import sent_tokenize, word_tokenize
sent_tokenized = []
for file in filenames:
    filename = './texts/'+file
    if '.DS_Store' in filename:
        continue
    else:
        with open(filename, 'r') as f:
            text = f.read()
            text = text.replace('\n', ' ')
            text = text.lower()
            #print(text)
            sent_tokenized.append(sent_tokenize(text))
#print(sent_tokenized[2][0:10])
for i in sent_tokenized:
    print(len(i))
```

```python
sent_tokenized = sent_tokenized[0] + sent_tokenized[1] + sent_tokenized[2]
print(sent_tokenized)
```

```python
from pymorphy2 import MorphAnalyzer
from nltk import word_tokenize
```

```python
def word_changer(sentence):
    #print(sentence)
    m = 'ruscorpora_mystem_cbow_300_2_2015.bin.gz'
    model = gensim.models.KeyedVectors.load_word2vec_format(m, binary=True, )
    morph = MorphAnalyzer()
    sentence_tok = [w.lower() for w in word_tokenize(sentence)]
    #print(sentence_tok)
    for i in range(len(sentence_tok)):
        a = morph.parse(sentence_tok[i])[0]
        value_dict = {}
        value_dict[i] = [a.tag.POS, a.tag.tense, a.tag.case, a.tag.number]
        try:
            if value_dict[i][0] == 'NOUN':
                similar = model.most_similar(a.normal_form+'_S')[0][0]
                similar = similar.split('_')[0]
                #print(similar)
                sentence_tok[i] = morph.parse(similar)[0].inflect({value_dict[i][2], value_dict[i][3]}).word
            if value_dict[i][0] == 'ADVB':
                similar = model.most_similar(a.normal_form+'_ADV')[0][0]
                similar = similar.split('_')[0]
                #print(similar)
                sentence_tok[i] = morph.parse(similar)[0].normal_form
        except Exception:
            continue
    sentence_tok = ' '.join(sentence_tok)
    sentence_tok = sentence_tok.replace(' ,', ',').replace(' .', '.').replace(' !', '!').replace(' ?', '?')
    return sentence_tok
```
